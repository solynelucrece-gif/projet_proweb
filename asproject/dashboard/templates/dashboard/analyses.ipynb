## ANALYSES STATISTIQUES DE LA BASE

```python
# importation des librairies

from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import openpyxl as pxl
```

```python
# importation de la base de donnes
data = pd.read_excel('C:\projet_proweb\data.xlsx')
#data.head()
sns.pairplot(data)
plt.show()
```

# ANALYSE UNIVARIEE

```python
#genre
plt.pie(data["gender"].value_counts(), labels=data["gender"].unique(), autopct='%1.1f%%')
plt.title("Rpartition par genre")
plt.show()
```

```python
# country
plt.pie(data["country"].value_counts(), labels=data["country"].unique(), autopct='%1.1f%%')
plt.title("Rpartition par pays")
plt.show()
```

```python
#age
plt.hist(data["age"], bins=10, edgecolor='black')
plt.title("Rpartition par ge")
plt.xlabel("ge")
plt.ylabel("Nombre de personnes")
plt.show()
```

```python
#niveau de revenu
plt.hist(data["income_level"], bins=10, edgecolor='black')
plt.title("Rpartition par niveau de revenu")
plt.xlabel("Niveau de revenu")
plt.ylabel("Nombre de personnes")
plt.show()
```

```python
#employment status
plt.pie(data["employment_status"].value_counts(), labels=data["employment_status"].unique(), autopct='%1.1f%%')
plt.title("Rpartition par statut d'emploi")
plt.show()  
```

```python
#relationship_status
sns.countplot(x="relationship_status", data=data)
plt.title("Rpartition par statut relationnel")
plt.xlabel("Statut relationnel")
plt.ylabel("Nombre de personnes")
plt.show()
```

```python
#has_children
plt.pie(data["has_children"].value_counts(), labels=data["has_children"].unique(), autopct='%1.1f%%')
plt.title("Rpartition par statut parental")
plt.show()
```

```python
# weekly_hours
plt.hist(data["weekly_work_hours"], bins=10, edgecolor='black')
plt.title("Rpartition par nombre d'heures de travail hebdomadaire")
plt.xlabel("Nombre d'heures de travail hebdomadaire")
plt.ylabel("Nombre de personnes")
```

```python

# Configuration du style global
sns.set_theme(style="whitegrid")

# Cration de la grille (2 lignes, 2 colonnes)
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Pie Chart : Subscription Status (Garder Matplotlib)
axes[0, 0].pie(data["subscription_status"].value_counts(), 
               labels=data["subscription_status"].unique(), 
               autopct='%1.1f%%', colors=sns.color_palette("pastel"))
axes[0, 0].set_title("Statut d'abonnement")

# 2. Pie Chart : Privacy Setting Level (Garder Matplotlib)
axes[0, 1].pie(data["privacy_setting_level"].value_counts(), 
               labels=data["privacy_setting_level"].unique(), 
               autopct='%1.1f%%', colors=sns.color_palette("muted"))
axes[0, 1].set_title("Niveau de confidentialit")

# 3. Bar Chart : Content Type Preference (Utiliser Seaborn countplot)
# countplot compte automatiquement les occurrences, plus besoin de .value_counts()
sns.countplot(data=data, x="content_type_preference", ax=axes[1, 0], palette="viridis")
axes[1, 0].set_title("Prfrence de type de contenu")
axes[1, 0].set_xlabel("Type de contenu")
axes[1, 1].set_ylabel("Nombre d'utilisateurs")

# 4. Bonus : Distribution ou autre variable (Exemple : Histogramme)
# On utilise l'espace vide axes[1, 1] pour une autre vue
sns.histplot(data=data, x="age", kde=True, ax=axes[1, 1], color="skyblue")
axes[1, 1].set_title("Distribution de l'ge")

# Ajustement automatique de l'espacement
plt.tight_layout()
plt.show()
```
